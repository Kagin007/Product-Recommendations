## Support
-Likelihood of all items in the rule being in a single transaction

## Confidence
-Likelihood of second item being bought once the first is bought

## Lift
-Change in likelihood of second item being bought once the first is bought

# Market Basket Analysis

How to measure the strength of a rule? A rule being 'if you buy A you will likely buy B'
# Measuring Strength
## Support 
- What proportion of all transactions contain both items? P(A, B) = 1% (1% is low indicating no correlation)
## Confidence
- Confidence = Supp(A, B) / Supp(B)
## Lift
- How much does the likelihood of buying A increase when B is bought?
